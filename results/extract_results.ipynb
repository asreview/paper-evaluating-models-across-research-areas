```python
# 28.04.2020
# This script is created to extract statistics from the simulation output files.
# it contains a function to extract results over 15 runs at once (extractseed) 
# and a function to extract results over 15 runs separately (extract15)
import json
import numpy as np
from asreviewcontrib.statistics import StateStatistics
import os 
```

```python
# define function to convert numpy int64 to integers
def convert_int64(mydict): 
    for key, value in mydict.items():
        if isinstance(value, np.int64):
            mydict[key] = int(value)
        elif isinstance(value, dict):
            convert_int64(value)
```

```python
# now for all seeded ones as well
def extractseed(path):
    # extract results
    state_dict = StateStatistics.from_path(path, wss_vals=[95, 99, 100], rrf_vals=[5, 10, 20]).to_dict()
    convert_int64(state_dict)
    
    # save results 
    out = "/Volumes/Gerbrich/asreview/results/one_seed/statistics/" + path.partition("one_seed/")[2] + "all.json"
    print(out)
    
    with open(out, "w") as f:
        json.dump(state_dict, f)
    
    return
```

```python
for root, dirs, files in os.walk("/Volumes/Gerbrich/asreview/simoutput/one_seed/", topdown=False):

   #for name in files:
    #  print(os.path.join(root, name))
    for name in dirs:
        #print(os.path.join(root, name))
        print(os.path.join(root,name) + "/")
        p = os.path.join(root,name) + "/"
        
        try:
        # block raising an exception
            extractseed(p)
        
        except:
            print("failed " + p)
            pass # doing nothing on exception
```

```python
# define new function that extracts results over all 15 runs for seeds
def extract15(path):
    # extract results
    
    state_dict = StateStatistics.from_path(path, wss_vals=[95, 99, 100], rrf_vals=[5, 10, 20]).to_dict()
    convert_int64(state_dict)
    
    fname = path.partition("one_seed/")[2]
    fname = fname.partition(".h5")[0] + ".json"
    #fname = fname.partition("results")[0] + "allruns/run"+ fname.partition(".h5")[0] + ".json"
    
    # save results 
    out = "/Volumes/Gerbrich/asreview/results/one_seed/statistics/" + fname
    print(out)
    
    with open(out, "w") as f:
        json.dump(state_dict, f)
    
    return
```
