```python
# function to create custom inclusion plot
def inclusion_plot(plot, output=None, **kwargs):
    all_files = all(plot.is_file.values())
    inc_plot = plot.new("inclusion", **kwargs)
    inc_plot.set_grid()
    inc_plot.set_xlim(0, 40)
    inc_plot.set_ylim(0, 101)
    for key in list(plot.analyses):
        if all_files or not plot.is_file[key]:
            inc_plot.add_WSS(key, 95, add_text = False)
            inc_plot.add_RRF(key, 10, add_text = False)
    inc_plot.add_random(add_text = False)
    inc_plot.set_legend()
    if output is None:
        inc_plot.show()
    else:
        inc_plot.show()
        inc_plot.save(output)
        inc_plot.close()
```

```python
# function to create custom inclusion plot, with no legend. 
def inclusion_plot_nolegend(plot, output=None, **kwargs):
    all_files = all(plot.is_file.values())
    inc_plot = plot.new("inclusion", **kwargs)
    inc_plot.set_grid()
    inc_plot.set_xlim(0, 40)
    inc_plot.set_ylim(0, 101)
    for key in list(plot.analyses):
        if all_files or not plot.is_file[key]:
            inc_plot.add_WSS(key, 95, add_text = False)
            inc_plot.add_RRF(key, 10, add_text = False)
    inc_plot.add_random(add_text = False)
    if output is None:
        inc_plot.show()
    else:
        inc_plot.show()
        inc_plot.save(output)
        inc_plot.close()
```
